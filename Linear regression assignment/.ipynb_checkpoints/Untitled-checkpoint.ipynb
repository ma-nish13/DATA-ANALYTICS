# Problem Statement
A bike-sharing system is a service in which bikes are made available for shared use to individuals on a short term basis for a price or free. Many bike share systems allow people to borrow a bike from a "dock" which is usually computer-controlled wherein the user enters the payment information, and the system unlocks it. This bike can then be returned to another dock belonging to the same system.


A US bike-sharing provider BoomBikes has recently suffered considerable dips in their revenues due to the ongoing Corona pandemic. The company is finding it very difficult to sustain in the current market scenario. So, it has decided to come up with a mindful business plan to be able to accelerate its revenue as soon as the ongoing lockdown comes to an end, and the economy restores to a healthy state. 


In such an attempt, BoomBikes aspires to understand the demand for shared bikes among the people after this ongoing quarantine situation ends across the nation due to Covid-19. They have planned this to prepare themselves to cater to the people's needs once the situation gets better all around and stand out from other service providers and make huge profits.


They have contracted a consulting company to understand the factors on which the demand for these shared bikes depends. Specifically, they want to understand the factors affecting the demand for these shared bikes in the American market. The company wants to know:

Which variables are significant in predicting the demand for shared bikes.
How well those variables describe the bike demands
Based on various meteorological surveys and people's styles, the service provider firm has gathered a large dataset on daily bike demands across the American market based on some factors. 



## Business Goal:
You are required to model the demand for shared bikes with the available independent variables. It will be used by the management to understand how exactly the demands vary with different features. They can accordingly manipulate the business strategy to meet the demand levels and meet the customer's expectations. Further, the model will be a good way for management to understand the demand dynamics of a new market.

## Creating Dummy variables(using one hot encoder method)
We will create dummy variables for following 4 categorical variables:<br>
1. Month<br>
2. Season<br>
3. Weekday<br>
4. Weather Situation

```python
bdata.head()
```

Refactoring values of columns Month, Season, Weekday and Weather Situation

```python
bdata['season'] = bdata['season'].map({1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'})
bdata.mnth = bdata.mnth.apply(lambda x: datetime.date(1900, x , 1).strftime('%B'))
bdata['weekday'] = bdata.weekday.map({0: 'sunday', 1: 'monday', 2: \
                                      'tuesday', 3: 'wednesday', 4: 'thursday', 5: 'friday', 6: 'saturday'})
bdata['weathersit'] = bdata.weathersit.map({1: 'clear', 2: 'mist', 3: 'light', 4: 'heavy'})
bdata.head()
```

```python
bdata_new = pd.get_dummies(bdata[['mnth', 'season', 'weekday', 'weathersit']], drop_first= True)
bdata_new.shape
```

```python
bdata = pd.concat([bdata, bdata_new], axis = 1)
```

```python
bdata = bdata.drop(columns = ['mnth','season','weekday','weathersit'])
bdata.head()
```

```python
bdata.shape
```

## Data Splitting

We will be normalizing following numerical columns before proceeding<br>

2. Temperature<br>

```python
#Initializing minmaxscalar
scaler = MinMaxScaler()
num_variable = ['dteday', 'temp', 'hum','windspeed', 'cnt']
df_train[num_variable] = scaler.fit_transform(df_train[num_variable])
df_train.head()
```

### Using correlation co-efficient

## Defining Target and Feature Variables

Training Target variable

```python
y_train = df_train.pop('cnt')
```

Training Feature(s) variable

```python
X_train = df_train
```

```python
X_train.shape
```

## Using Recursive Feature Elimination(RFE)
We will use RFE which provides us the set of variables ranked in order of importance of their impact on outcome
We will choose 15 variables for our analysis

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm, 15)
rfe = rfe.fit(X_train, y_train)
```

```python
rfe_results = pd.DataFrame({'Variable': X_train.columns, 'Status': rfe.support_, 'Rank': rfe.ranking_})
rfe_results.sort_values(by='Rank')
```

## Model Building

We will now build the model manually by validating these 15 features and pruning any feature that is not satisfactory

```python
columns = X_train.columns[rfe.support_]
X_train_rfe = X_train[columns]
X_train_sm = sm.add_constant(X_train_rfe)
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

## VIF Calculation
We will make a function to properly depict VIF values

```python
def vif_res(local_df):
    vif_df = pd.DataFrame()
    vif_df['Features'] = local_df.columns
    vif_df['VIF'] = [variance_inflation_factor(local_df.values, i) for i in range(local_df.shape[1])]
    vif_df['VIF'] = round(vif_df['VIF'], 2)
    vif_df = vif_df.sort_values(by='VIF', ascending=False)
    vif_df = vif_df.reset_index(drop=True)
    return vif_df
```

```python
# obtaining VIF values
vif_res(X_train_rfe)
```

Looking at the p-values and VIF, giving preference to p-values, we find out<br>
The p-value for Holiday feature is greater than 0.05(its 0.168), hence we move ahead and drop this feature and redesign our model

```python
columns = columns.drop(['holiday'])
X_train_rfe = X_train[columns]
X_train_sm = sm.add_constant(X_train_rfe)
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

```python
# getting the VIF values
vif_res(X_train_rfe)
```

From above analysis, we find out the p-values are all below the signifance level of 0.05, but we see some VIF values being greater than 5, so now we will move ahead with our analysis based on VIF values

#### Dropping humidity feature and redesigning

```python
columns = columns.drop(['hum'])
X_train_rfe = X_train[columns]
X_train_sm = sm.add_constant(X_train_rfe)
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

```python
# Obtaining VIF values
vif_res(X_train_rfe)
```

From above analysis, p-values are under control, but we see working day column having VIF of 11 approx.<br>
We will drop workingday and redesign the model

#### Dropping workingday column

```python
columns = columns.drop(['workingday'])
X_train_rfe = X_train[columns]
X_train_sm = sm.add_constant(X_train_rfe)
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

```python
#obtaining VIF values
vif_res(X_train_rfe)
```

We now see that p-value of 'weekday_saturday' is more than 0.05, we will drop it and redesign

#### Dropping 'weekday_saturday' and redesigning the model

```python
columns = columns.drop(['weekday_saturday'])
X_train_rfe = X_train[columns]
X_train_sm = sm.add_constant(X_train_rfe)
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

```python
vif_res(X_train_rfe)
```

We decide to keep the 'temp' feature as the value is not significantly higher than 5

#### Now we can conclude that the P-values and VIF are under control and go with further analysis of the model

## Proof for Linear Regression Assumptions

### Normally distributed error terms check

```python
y_train_pred = lr_model.predict(X_train_sm)
res = y_train_pred - y_train
sns.distplot(res)
plt.xlabel('Residual of Error Terms', fontdict= {'fontsize': 22})
plt.title('Distribution of Residual Errors', fontdict= {'fontsize': 22})
plt.show()
```

#### We can see that the residual errors are normally distributed with mean value almost 0, implying the model is significant

### Error terms independence check

```python
sns.regplot(x = y_train_pred, y = res)
plt.xlabel('Predicted Values', fontdict = {'fontsize': 18})
plt.ylabel('Residual errors', fontdict = {'fontsize': 18})
plt.title('Residual errors vs predict values', fontdict = {'fontsize': 18})
plt.show()
```

#### We see from above graph that there is no pattern or dependence between Residuals and Predicted Values which was expected from a significant model.

### Checking Homoscedasticity

```python
sns.regplot(x=y_train, y=y_train_pred)
plt.xlabel('Actual Points', fontdict={'fontsize': 18})
plt.ylabel('Predicted Points', fontdict={'fontsize': 18})
plt.title('Predicted Points Vs. Actual Points', fontdict={'fontsize': 22})
plt.show()
```

#### Above graph proves the Homoscedasticity of error terms

### Evaluation of correlation coefficient of feature variables with target variable

```python
from scipy.stats import pearsonr
corrd= {index: round(pearsonr(y_train, X_train_rfe[index])[0], 2) for index in X_train_rfe.columns}
corr_df = pd.DataFrame(corrd.values(), index=corrd.keys(), columns=['Correlation_Coef'])
corr_df.iloc[(-corr_df['Correlation_Coef'].abs()).argsort()]
```
